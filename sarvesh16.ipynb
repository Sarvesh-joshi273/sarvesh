```python

sns.histplot(data['age'])
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

sns.histplot(data['bmi'])
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()

sns.histplot(data['children'])
plt.title("Distribution of Children")
plt.xlabel("Number of Children")
plt.ylabel("Frequency")
plt.show()


sns.histplot(data['expenses'])
plt.title("Distribution of Charges")
plt.xlabel("Charg")
plt.ylabel("Frequency")
plt.show()
```

```python

sns.countplot(x='sex', data=data)
plt.title("Count of Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()

sns.countplot(x='smoker', data=data)
plt.title("Count of Smoker")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.show()

sns.countplot(x='region', data=data)
plt.title("Count of Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show()
```

```python
num_data = data[['age', 'bmi', 'children', 'expenses']]

corr = num_data.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Average charge:", data['expenses'].mean())
print("Maximum charge:", data['expenses'].max())
print("Minimum charge:", data['expenses'].min())
print("Average charges by smoker status:\n", data.groupby('smoker')['expenses'].mean())
```

```python
sns.boxplot(x='smoker', y='expenses', hue='sex', data=data)
plt.title("Charges by Smoker and Sex")
plt.show()
```
